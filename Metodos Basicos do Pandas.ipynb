# Adicionando uma nova coluna na tabela

# Excluir linhas e colunas

```python
 #1 eixo(axis) da coluna, 0 é o da linha
#vendaDf = vendaDf.drop("Imposto", axis=1)       
#display(vendaDf)
vendaDf = vendaDf.drop(0, axis=0)
display(vendaDf)
```

# Valores vazios(NaN)

```python
#deletar linhas e colunas completamentes vazias
#vendaDf = vendaDf.dropna(how='all', axis=1)        #Exclui todos os valores vazios

#Valor vazio avulso em alguma linha, se tiver pelo menos 1 valor vazio nessa tal linha:
#vendaDf = vendaDf.dropna()

#Preencher valores vazios
#Preenche com a media da coluna(media de uma coluna -> vendaDf['Comissão'].mean()):

vendaDf['Comissao'] = vendaDf['Comissao'].fillna(vendaDf['Comissao'].mean())
display(vendaDf)

#Preencher uma linha sem nada com o ultimo valor de cima
vendaDf = vendaDf.ffill()
```

# Calcular indicadores
-Groupby
-Value Counts

```python
#transacaoLoja = vendaDf['ID Loja'].value_counts()
#display(transacaoLoja)


```

```python
#Agrupar informações
#faturaProduto = vendaDf.groupby('Produto').sum()
#display(faturaProduto)

#coluna especifica(com  soma .sum(), media .mean())
faturaProduto = vendaDf[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturaProduto)
```

# Mesclar 2 dataframes, procurar informações de um dataFrame em outro

```python
gerentesDf = pd.read_excel("Gerentes.xlsx")
#display(gerentesDf)

#Colocar o gerente de cada loja, em sua devida linha, na tabela original

vendaDf = vendaDf.merge(gerentesDf)
display(vendaDf)
```
